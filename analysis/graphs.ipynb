```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
grid_size_data = pd.read_csv('../analysis/results/grid_size_results.csv')
results = grid_size_data.groupby('x')['final_cycle'].agg(['mean', 'std']).reset_index()
results.columns = ['x', 'final_cycle_mean', 'final_cycle_std']

# Create the plot
plt.figure(figsize=(10, 6))

# Plot mean with error bars (standard deviation)
plt.errorbar(results['x'], results['final_cycle_mean'], 
             yerr=results['final_cycle_std'], 
             marker='o', capsize=5, capthick=2, 
             linewidth=2, markersize=8,
             label='Mean ± Std Dev', color='#A23B72')

plt.xlabel('x', fontsize=12)
plt.ylabel('Moves', fontsize=12)
plt.title('Moves vs x (Mean ± Std Dev)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()

# Save the plot
plt.savefig('../analysis/1.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
# Your existing code
weight_data = pd.read_csv('../analysis/results/weight_results.csv')
aggregated = weight_data.groupby('weight').agg({
    'won': 'mean',
    'final_cycle': ['mean', 'std']
})
aggregated.columns = ['success_rate', 'final_cycle_mean', 'final_cycle_std']
aggregated = aggregated.reset_index()
print(aggregated)

# Create two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Success Rate vs Weight
ax1.plot(aggregated['weight'], aggregated['success_rate'], 
         marker='o', linewidth=2, markersize=8, color='#2E86AB')
ax1.set_xlabel('Weight', fontsize=12)
ax1.set_ylabel('Success Rate', fontsize=12)
ax1.set_title('Success Rate vs Weight', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0, 1.05])  # Success rate is typically 0-1

# Plot 2: Final Cycle (mean with error bars) vs Weight
ax2.errorbar(aggregated['weight'], aggregated['final_cycle_mean'], 
             yerr=aggregated['final_cycle_std'], 
             marker='o', linewidth=2, markersize=8, 
             capsize=5, capthick=2, color='#A23B72')
ax2.set_xlabel('Weight', fontsize=12)
ax2.set_ylabel('Moves', fontsize=12)
ax2.set_title('Moves vs Weight (Mean ± Std)', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../analysis/3.png', dpi=300, bbox_inches='tight')
plt.show()
```
